```python
df_dsn = df_can.loc[['Sweden','Denmark','Norway'],:]
```

```python
df_dsn.head()
```

```python
total_value = df_dsn[['Total']].sum()
category_proportions=[(value / total_value) for value in df_dsn['Total']]
```

```python
for i,prop in enumerate(category_proportions):
    print(df_dsn.index.values[i],prop)
```

```python
width=40
height=10
total_num_tiles=width*height
tiles_per_category = [round(val*total_num_tiles) for val in category_proportions]
```

```python
waffle = np.zeros((width,height))
tile_count=0.0
category_index = 1

    
```

```python
for i in range(width):
    for j in range(height):    
        tile_count+=1
        if tile_count > float(sum(tiles_per_category[0:category_index])):
            category_index+=1
        waffle[i,j]=category_index
        
```

```python
from PIL import Image
```

```python
plt.matshow(waffle,cmap='coolwarm')
#plt.colorbar()

ax = plt.gca()
#ax.set_xticks(np.arange(-.5,(width),1))
#ax.set_yticks(np.arange(-.5,(height),1))
ax.grid(which='minor',linestyle='-',linewidth=2,color='w')
plt.xticks([])
plt.yticks([])
```

```python
df_incidents = pd.read_csv('https://ibm.box.com/shared/static/nmcltjmocdi8sd5tk93uembzdec8zyaq.csv')
limit = 100
df_incidents = df_incidents.iloc[0:limit, :]
```

```python
# San Francisco latitude and longitude values
latitude = 37.77
longitude = -122.42


# In[ ]:


# create map and display it
sanfran_map = folium.Map(location=[latitude, longitude], zoom_start=12)
incidents = folium.map.FeatureGroup()

# loop through the 100 crimes and add each to the incidents feature group
for lat, lng, in zip(df_incidents.Y, df_incidents.X):
    incidents.add_child(
        folium.features.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6          
        )
    )

# add incidents to map
sanfran_map.add_child(incidents)
```

```python
latitudes = list(df_incidents.Y)
longitudes = list(df_incidents.X)
labels = list(df_incidents.Category)

for lat, lng, label in zip(latitudes, longitudes, labels):
    folium.Marker([lat, lng], popup=label).add_to(sanfran_map)  
```

```python
df_incidents = pd.read_csv('https://ibm.box.com/shared/static/nmcltjmocdi8sd5tk93uembzdec8zyaq.csv')
df_incidents = df_incidents.groupby('PdDistrict', as_index=False)
```

```python
df_incidents = df_incidents.count()
```

```python
world_geo = r'san-francisco.geojson' # geojson file

# create a plain world map
latitude = 37.77
longitude = -122.42
world_map = folium.Map(location=[latitude, longitude], zoom_start=12)

world_map.choropleth(
    geo_data=world_geo,
    data = df_incidents,
    columns = ['PdDistrict', 'IncidntNum'],
    key_on = 'feature.properties.DISTRICT',
    fill_color='YlOrRd', 
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='Crime Rate in San Francisco'
)

# display map
world_map
```
